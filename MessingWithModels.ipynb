### Create an SVM that given the ingredients, predict indian or not indian

```python
exercise_dataset = indianFood[['TranslatedIngredients', 'Cuisine']]
```

```python
exercise_dataset #remove outliers, not translated text
```

```python
import langdetect as ld

save_rows = []

for i, row in enumerate(exercise_dataset.TranslatedIngredients):
    try:
        if ld.detect(row) != 'en':
            save_rows.append(i)
    except:
        save_rows.append(i)
        continue
```

```python
exercise_dataset = exercise_dataset.drop(exercise_dataset.index [save_rows])
```

```python
exercise_dataset
```

```python
import matplotlib.pyplot as plt
```

```python
import json
from collections import defaultdict
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
stop_words = set(stopwords.words('english'))
nltk_tokenize = lambda text: [re.sub(r'\d+', '', x.lower()) for x in nltk.word_tokenize(text) if x not in stop_words] #solo if alla fine
```

```python
#need doc * term matrix
#need words associated to a specific cuisine
cuisine_frequency = defaultdict(lambda: defaultdict(lambda: 0))

for recipe, cuisine in exercise_dataset.itertuples(index=False):
    for ingredient in nltk_tokenize(recipe):
        cuisine_frequency[cuisine][ingredient] += 1
        
cuisine_frequency
```

```python
#QUANTO SONO USATI GLI INGREDIENTI ITALIANI VS QUELLY INDIANI?
italian = cuisine_frequency['Italian Recipes']
indian = cuisine_frequency['Indian']
```

```python
common_ingredients = [x for x in list(italian.keys()) if x in list(indian.keys())]
values_indian = [indian[i] for i in common_ingredients]
values_italian = [italian[i] for i in common_ingredients]
plt.scatter(values_indian, values_italian)
```

```python
#try to plot using tdidf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
# list of text documents
corpus = []
for row in exercise_dataset.TranslatedIngredients:
    corpus.append(row)

# create the transform
vectorizer = TfidfVectorizer()

# tokenize and build vocab
tfidf = vectorizer.fit_transform(corpus)
tfidf.shape
```

```python
tfidf[1]
```

```python
#SVM : I need for each recipe, a vector that tells me the frequency of words, the vector is composed by all possible words, and a label 1, 0 for indian or not indian, other ideas?
#I could use the tfidf matrix instead of just using the frequency of elements
```

```python
tfidf_dataframe = pd.DataFrame(tfidf.todense(), columns= vectorizer.get_feature_names()) #rows = recipes aka documents, colums = words
```

```python
tfidf_dataframe
```

```python
rec = tfidf_dataframe.iloc[0, :] #all elements of first recipe

for i in rec:
    print(i) #as we can see, some values are higher than 0, so the tfidf matrix seems ok
```

```python
# I'd like to add a column to the dataset used for prediction purposes, 1 if it's indian or indian related, 0 otherwise
GT = [1 if 'Indian' in x else 0 for x in exercise_dataset.Cuisine] #ground_truth
GT
```

```python
tfidf_dataframe['ground_truth'] = GT
tfidf_dataframe
```

```python
X = tfidf_dataframe.iloc[:, 0:-2].values    #features
y = tfidf_dataframe.iloc[:, -1].values      #labes
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) #creating training set and test set
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
```

```python
predictions = svm.predict(X_test)
print(predictions[:20])
```

```python
print(y_test[:20])
```

```python
#just to get handy
TP, TN, FP, FN = 0,0,0,0
for i, pred in enumerate(predictions):
    if pred == 1 and pred == y_test[i]:
        TP +=1
    elif pred == 0 and pred == y_test[i]:
        TN +=1
    elif pred == 1 and pred != y_test[i]:
        FP +=1
    else:
        FN +=1
        
print(TN, FP, FN, TN)
```

```python
from sklearn.metrics import confusion_matrix

#Generate the confusion matrix
cf_matrix = confusion_matrix(y_test, predictions)

print(cf_matrix) #exactly as the ones that I counted by myself!
```

```python
import seaborn as sns

ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
Precision = TP/(TP + FP)
Recall = TP/(TP + FN)

print(Precision, Recall)
```
